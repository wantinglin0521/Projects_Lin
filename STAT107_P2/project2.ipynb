### Puzzle 4.1

```python
# Analyze the average ranking of the restaurants in the United States appear in the list form 2002 to 2023. 

# Grouping the dataframe.
rank_sum_US = df[df["country"] == "United States"].groupby("year").agg("sum", numeric_only=True).reset_index()
rank_count_US = df[df["country"] == "United States"].groupby("year").agg("count").reset_index()

# Calculate the average ranking.
rank_sum_US["avg"] = rank_sum_US["rank"] / rank_count_US["rank"]
rank_sum_US.rename(columns={"rank": "sum"}, inplace=True)
rank_sum_US[["year", "sum", "avg"]]
```

### Puzzle 4.2

```python
# Analyze the average ranking of the restaurants in the France appear in the list form 2002 to 2023. 

# Grouping the dataframe.
rank_sum_FR = df[df["country"] == "France"].groupby("year").agg("sum", numeric_only=True).reset_index()
rank_count_FR = df[df["country"] == "France"].groupby("year").agg("count").reset_index()

# Calculate the average ranking.
rank_sum_FR["avg"] = rank_sum_FR["rank"] / rank_count_FR["rank"]
rank_sum_FR.rename(columns={"rank": "sum"}, inplace=True)
rank_sum_FR[["year", "sum", "avg"]]
```

## Section 5: Data Science

By observing the result from Puzzle 2 plot, I found out that the U.S. and France are the first and second countries that have the most World‚Äôs 50 Best Restaurants from 2002 to 2023. 

I would like to know which country has better average ranking. Hence, I plotted out the average ranking of the restaurants in the U.S. and in France to compare the result. However, sometimes U.S. wins, while sometimes France wins. This shows that they are highly comparable and this also intrigues me to find out which country has a better average ranking from 2002 to 2023. As a result, I decide to do a hypothesis testing on the above mention statement.

### Defining Our Hypotheses

*$H_0: \mu _{US} >= \mu _{FR}$*

*$H_1: \mu _{US} < \mu _{FR}$*

Where $\mu _{US}$ refers to the average ranking of the US restaurants
and $\mu _{FR}$ refer to the average ranking of the French restaurants.

## Puzzle 5.1: Prepare for the Dataframe

```python
# Filter out all restaurants which are located in the U.S. in the dataset from 2002 to 2023.
sample_US = df[(df["country"] == "United States")]

# Test code.
#sample_US
```

```python
# Filter out all restaurants which are located in France in the dataset from 2002 to 2023.
sample_FR = df[(df["country"] == "France")]

# Test code.
#sample_FR
```

## Puzzle 5.2: Conducting the Hypothesis Test

```python
# Hypothesis test.
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(sample_US["rank"], sample_FR["rank"], equal_var=True)

print(f"T-Statistic: {round(t_stat, 4)}")
print(f"P-Value: {round(p_value, 4)}")
```

### Observation for Puzzle 5.2

> *"p-value is 0.731 which is higher than 0.05(siginficant level). Therefore, we do not reject the null hypothesis, meaning the U.S. does not have the better restaruants than France."*

## Section 6: Different Data Visualization

In order to observate the differences between the average ranking of the U.S. and France, I decided to visualize the box plot of the them.

In the plot, the green triangle shows the mean and the orange line shows the median.

```python
# Plot the side by side Box plot.
# Prepare data for plotting.
US_rankings = df[df["country"] == "United States"]["rank"]
FR_rankings = df[df["country"] == "France"]["rank"]

plt.boxplot([US_rankings, FR_rankings], labels=["US", "France"], showmeans=True)

# Adding labels and title.
plt.title("Restaurant Rankings by Countries")
plt.ylabel("Ranking")
plt.xlabel("Country")
plt.grid()
```

### Observation for Section 6

> *"The plot shows that there is no huge difference between the U.S. and French restaurants in terms of ranking."*
